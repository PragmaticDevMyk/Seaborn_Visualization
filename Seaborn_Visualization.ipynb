```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
```

```python
cancer_df = pd.read_csv('cancer.csv')
cancer_df.head()
```

```python
# Set plot style
sns.set(style="whitegrid")

# Create scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=cancer_df, x="mean area", y="mean smoothness", hue="target", palette="Set1", alpha=0.7)

# Set titles and labels
plt.title("Scatter Plot: Mean Area vs. Mean Smoothness")
plt.xlabel("Mean Area")
plt.ylabel("Mean Smoothness")
plt.legend(title="Target")
plt.tight_layout()
plt.show()
```

```python
# Create a count plot for the target classes
plt.figure(figsize=(6, 5))
sns.countplot(data=cancer_df, x="target", palette="Set2")

# Set titles and labels
plt.title("Count of Samples by Class")
plt.xlabel("Target Class")
plt.ylabel("Count")
plt.xticks(ticks=[0, 1], labels=["Class 0", "Class 1"])
plt.tight_layout()
plt.show()
```

```python
# Select relevant columns
selected_columns = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness']
df_selected = cancer_df[selected_columns + ['target']]

# Define color mapping for target classes
colors = df_selected['target'].map({0: 'red', 1: 'blue'})

# Create scatter matrix
plt.figure(figsize=(12, 12))
scatter_matrix(df_selected[selected_columns], figsize=(12, 12), diagonal='kde', color=colors, alpha=0.6)

plt.suptitle("Scatter Matrix of Selected Features", fontsize=16)
plt.show()
```

```python
#Plot a heatmap showing the correlation between all features
# Compute the correlation matrix
correlation_matrix = cancer_df.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(16, 14))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, square=True, cbar_kws={"shrink": .5})

# Add title
plt.title("Feature Correlation Heatmap", fontsize=18)
plt.tight_layout()
plt.show()
```
